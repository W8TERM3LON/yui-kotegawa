```
import requests
import json, os, re

url = "https://graphql.anilist.co/"


# To get access token please visit: https://anilist.co/api/v2/oauth/authorize?client_id=2861&response_type=token
accessToken = 'REPLACEWITHOAUTH2TOKEN'
accessToken = accessToken.replace('https://anilist.co/user/W8TERM3LON#access_token=','')
accessToken = re.sub("&token_type.*$", "", accessToken)

#user specified page number
no = input('Enter page number --->')

#POST request data
payload = "{\"query\":\"query ($p: Int) {\\n  Page(page: $p) {\\n    pageInfo {\\n      total\\n      perPage: currentPage\\n      lastPage\\n      hasNextPage\\n    }\\n    users(sort: ID_DESC) {\\n      id\\n      avatar{\\n        large\\n      }\\n      bannerImage\\n      \\n      \\n    }\\n  }\\n}\",\"variables\":{\"p\":" + str(no) + "}}"
headers = {
    'content-type': "application/json",
    'authorization': "Bearer " + accessToken
    }

#POST request parameters.
response = requests.request("POST", url, data=payload, headers=headers)

#Output - query data
qdata = response.text
#remove backslashes.
qdata = qdata.replace('\\','')
#remove square brackets.
qdata = qdata.replace('[', '')
qdata = qdata.replace(']', '')
#Remove default image profiles.
qdata = qdata.replace('"https://s4.anilist.co/file/anilistcdn/user/avatar/large/default.png"', 'null')
#remove "large":"
qdata = qdata.replace('"large":"','')
#remove quotation marks
qdata = qdata.replace('"','')

#DO NOT CASEFOLD. IMAGE URL IS CASE SENSITIVE
#Turn qdata into an array at each },{
rdata = qdata.split('},{')


#loop 50 times with +1 each time
nq = 0
while nq < 49:
    #set image as rdata
    image = rdata[nq]
    idnum = rdata[nq].split(',')
    idnum = idnum[0].replace('id:','')
    print('https://anilist.co/user/' + idnum)
    nq = nq+1
    match = re.search("{(.+?)}", image,flags=re.IGNORECASE)
    result = match.group(1)
    result = result.replace('large:null','')
    print(result)
```
